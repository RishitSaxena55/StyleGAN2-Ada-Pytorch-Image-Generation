```python
from PIL import Image
import os
import math

def create_image_collage(folder_path, output_path="collage.png", max_images=None, bg_color="white"):
    """
    Create a collage from images in the folder with automatic grid sizing.

    Parameters:
    - folder_path: str - Path to folder with images
    - output_path: str - Output collage image path
    - max_images: int or None - Maximum number of images (optional)
    - bg_color: str - Background color (default 'white')
    """
    # Get all image files
    image_files = sorted([
        os.path.join(folder_path, f)
        for f in os.listdir(folder_path)
        if f.lower().endswith(('.png', '.jpg', '.jpeg', '.webp'))
    ])

    if not image_files:
        print("❌ No image files found in the folder.")
        return

    if max_images:
        image_files = image_files[:max_images]

    num_images = len(image_files)
    grid_cols = math.ceil(math.sqrt(num_images))
    grid_rows = math.ceil(num_images / grid_cols)

    images = [Image.open(p).convert("RGB") for p in image_files]

    # Resize images to match the first image size
    width, height = images[0].size
    images = [img.resize((width, height)) for img in images]

    # Create blank canvas
    collage = Image.new("RGB", (grid_cols * width, grid_rows * height), color=bg_color)

    for idx, img in enumerate(images):
        x = (idx % grid_cols) * width
        y = (idx // grid_cols) * height
        collage.paste(img, (x, y))

    collage.save(output_path)
    print(f"✅ Collage saved at: {output_path}")
```

## Data Preprocessing

```python
# !pip install face-alignment
import face_alignment
from PIL import Image
import numpy as np

fa = face_alignment.FaceAlignment(face_alignment.LandmarksType.TWO_D, device='cuda' or 'cpu')

def align_face(image_path, save_path='aligned.png'):
    input = Image.open(image_path).convert('RGB')
    input_np = np.array(input)
    preds = fa.get_landmarks(input_np)
    if preds is None:
        print("No face detected.")
        return
    # (Simple crop around center if full alignment is too complex)
    input = input.crop((50, 50, input.width - 50, input.height - 50)).resize((1024, 1024))
    input.save(save_path)
```
